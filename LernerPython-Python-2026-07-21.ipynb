# Strings use Unicode

It used to be that Python strings were really bytes. If you used just ASCII, you never felt any issues. But the moment you wanted to use other languages, or emojis, you got into trouble -- you needed to use "Unicode strings."

Now things have reversed: Strings only work with Unicode, and if you want bytes, you need a "bytestring"

```python
s = 'abcde'
len(s)
```

```python
s = 'אבגדה'
len(s)
```

```python
s[0]
```

```python
ord(s[0])  # this gives me the Unicode number, not ASCII
```

```python
import unicodedata
```

```python
unicodedata.name(s[0])
```

```python
l/   # L/ukas  -- Wukash 
```

```python
# what if I want to read in a binary file?
# Or what if I want to work with bytes?
# I need to use a bytestring

s
```

```python
b = s.encode() # this returns a bytestring
b
```

```python
b[0]
```

If you read a binary file into Python, you must read it with `'rb'` and you will not get a string, but rather a bytestring

```python
b[1:]
```

```python
b[1:].decode()  # let's turn this into a string
```

# Other big string news: f-strings!

These were introduced in Python 3.7. They replace two previous string interpolation systems:

- We could use `%` as an operator on strings -- this still exists, and is used in the `logging` module
- We can also use `str.format`, a method that lets us interpolate



```python
x = 10
y = 20
z = [10, 20, 30]

print(f'{x} + {y} = {x+y}')
```

```python
print(f'{z}')
```

```python
f'{x=}'  # this means: print the expression and its value
```

```python
f'{x = }' 
```

```python
f'{len(z)=}' 
```

```python
s = 'abcdefg'

print(f'{s!r}')  # this returns the repr of the object
```

```python
print(f'{z!r}')
```

```python
# repr == the object's internal representation, as shown to programmers in Jupyter and debuggers
```

```python
d = {'a':1000, 'bcdef':2, 'ghi':300}

for key, value in d.items():
    print(f'{key}: {value}')
```

Inside of the `{}` for the f-string, you can put a `:`. After the `:` comes a format specifier. Every type has its own format specifiers. For strings, you can indicate the size of the field you want to use.

```python
d = {'a':1000, 'bcdef':2, 'ghi':300}

for key, value in d.items():
    print(f'{key:.<10}{value:.>10}')
```

fstring help: https://fstring.help/

```python
n = 12345.67890

print(f'{n:.3f}')
```

```python
print(f'{n:3%}')
```
