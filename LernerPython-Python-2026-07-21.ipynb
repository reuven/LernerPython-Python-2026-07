# Structural pattern matching

For *YEARS*, people new to Python complained that it didn't have a switch-case statement. And then, Python added match-case, and said that you can use this instead of `if`. 

Except that it's not the same:

- `switch-case` in C is for `if`-`else` decision-making
- `match`-`case` in Python is for breaking apart data structures without having to work hard

````python
name = 'Reuven'

match name:
    case 'Reuven':
        print('Hi, boss!')
    case 'someone else':
        print('Hi, someone else')
    case _:
        print('This is weird')
````

Traditionally, the final `case` is put into a variable. That means: It'll match anything, anything at all. The `_` variable is used for throwaway data.

````python
name = 'Reasdfsadfasdfdasdfdadsuven'

match name:
    case 'Reuven':
        print('Hi, boss!')
    case 'someone else':
        print('Hi, someone else')
    case other_name:  # the variable here is assigned the value in the "match" statement!
        print('This is weird')
````

````python
other_name
````

````python
# structural pattern matching is really meant for picking apart values from a collection -- sort of like we do with unpacking

mylist = [10, 20]

match mylist:
    case [x]:
        print(f'{x=}')
    case [x, y]:
        print(f'{x=}, {y=}')
    case [x, y, z]:
        print(f'{x=}, {y=}, {z=}')
    case other:
        print(f'{other=}')
````

````python
# we can use our unpacking *z stuff here, too:

mylist = [10, 20, 30, 40, 50, 60]

match mylist:
    case [x]:
        print(f'{x=}')
    case [x, y]:
        print(f'{x=}, {y=}')
    case [x, y, *z]:
        print(f'{x=}, {y=}, {z=}')
    case other:
        print(f'{other=}')
````

````python
# we can match dicts, too:

d = {'a':10, 'b':20, 'c':30}

match d:
    case {'a':10, 'b':20}:
        print('Partial match')
    case _:
        print('Default')
````

````python
name = 'asdfsafdsdafsafsafsafsafas'

match name:
    case x if len(x) < 5:
        print(f'Short name, {name}!')
    case x if len(x) > 10:
        print(f'Long name, {name}!')
    case _: 
        print(f'Medium name, {name}!')
````

# Modern dicts

For many years, one of the most important things to tell people about dicts was that they didn't preserve their order. If you created `{'a':10, 'b':20, 'c':30}`, you didn't know if iterating would give you `a`, `b`, `c`, or maybe `b`, `c`, and `a`.

Then, in Python 3.6 and 3.7, they changed this entirely: In 3.6, they changed the implementation of dicts, and in 3.7 they said that a nice side feature (namely, the ordering) was guaranteed.

What did they change?

A dict is a key-value store, a hash table. This means that if you store `d['a'] = 10`, Python runs the `hash('a')`, gets a number, and (in old versions of Python) would use that number to know where to store your value. Iterating over a dict meant going through each of the keys, invoking `hash` on each key, and retrieving each value.

Not only did this not guarantee the order, but it also meant that the dict needed a *lot* of space in memory. If we didn't use lots of memory, then we would end up with many colissions, which is a problem in dicts.

In 3.6, they changed the implementation to use 2 data structures:

1. a table with three columns that grows over time. This is where the key and value are stored, sequentially. The third column is the index in the table, starting with 0. There's one row in this table for each key-value pair in the dict.
2. an array (at the C level), which is where the hash function is really invoked. This is much smaller than the original dict ever was, because it's just integers in C.

If you store `d['a'] = 10`, Python first puts a new row in the table with the index (automatically calculated) and then the key-value pair. Then Python calculates `hash('a') % len(array)`, and gets an indication of where to put the index from the table.

So the array contains table-row indexes. Each lookup in the dict requires two lookups -- one in the array and one in the dict.

But we save a TON of memory! 

Not only does this save memory, but we also have the table with our key-value pairs in chronological order!

What happens, though, when we remove a key-value pair? 

Remember that all variables in Python are actually keys in a dict. Remember that every attribute (i.e., names after `.`) are really dicts. Remember that objects are really dicts.

````python
d = {'a':10, 'b':20, 'c':30}

for key, value in d.items():
    print(f'{key}: {value}')
````

````python
d.pop('b')
d['b'] = 999

for key, value in d.items():
    print(f'{key}: {value}')
````

# The great de-listing

Back in Python 2, we had two different `range` functions:

- `range(n)`, which returned a list of integers from 0 to `n-1`
- `xrange(n)`, which returned an iterable object that didn't contain all of the numbers, but could generate them in a `for` loop.

Why use `range` in such a world? Just use `xrange` -- for small things, no one cares, and for big things, everyone cares!

In Python 3, they got rid of `range`, and replaced it with `xrange`, which they renamed to `range`. Basically, our modern `range` is the old `xrange`.

The larger story is that Python is now being used for large-scale data. The notion that we'll get lists back from things is a bit scary -- if the list is too big, it'll blow up our computer's memory and/or crash our program.

# Iterator protocol

Python is using generators more and more, and they are objects that know how to use the iterator protocol. This protocol is not new, but it's just everywhere now:

1. An iterable object should respond to the `iter` builtin function, returning its iterator.
    - If it's not iterable, then it raises an exception
2. An iterator should respond to the `nextr` builtin function, returning the next value it has to offer.
    - If it's at the end, then it raises the `StopIteration` exception

Behind the scenes, what does it mean for an object to know how to respond to `iter` or `next`?

The answer: It implements two methods.

- If an object implements the `__iter__` method, then it is considered iterable. This method should return an iterator. (It could return the object itself, if the object implements `__next__`.)
- If an object imlements the `__next__` method, then it is an iterator. This method will be invoked repeatedly by `next`, until we get a `StopIteration` exception, which stops the loop.

````python
# to make any class iterable, just implement those methods

class MyData:
    def __init__(self, data):
        self.data = data
        self.index = 0

    def __iter__(self):  # I'm my own iterator! -- this method is invoked once, at the start of the loop
        return self 

    def __next__(self):
        if self.index >= len(self.data):  # if we're at the end, indicate we're done -- this is invoked many times, once per iteration
            raise StopIteration
            
        value = self.data[self.index]
        self.index += 1
        return value

m = MyData('abcd')

for one_item in m:
    print(one_item)
````

````python
# we can instead have __iter__ return a generator, via a generator function

class MyData:
    def __init__(self, data):
        self.data = data

    def __iter__(self):  
        for one_item in self.data:
            yield one_item

m = MyData('abcd')

for one_item in m:
    print(one_item)
````

# Type hints

Python is a famously dynamically typed language. What does that mean?

1. Variables are not aliases to locations in memory. Variables in Python are references to objects -- every object is on the heap, and every reference just points to that location in memory.
2. This means that every variable is just an address i
