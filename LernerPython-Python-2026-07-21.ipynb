# Exercise: Format receipt

Given a list of 3-element tuples (product name, unit price, and quantity), print a receipt with names padded to a fixed width, prices to two decimals, and a computed total per row.

Write a function, `print_receipt`, that takes a list of such 3-element tuples and prints on the screen.

https://practice.lernerpython.com/classroom/bfb596ff6a/ex-1

```python
def print_receipt(items):
    # item is a bunch of tuples
    # each tuples will be name, price, quantity

    for one_item in items:
        name, price, quantity = one_item
        print(f'{name:10}\t{price:^6.2f}\t${quantity * price:.2f}')

items = [
    ('Widget', 9.99, 3),
    ('Gadget', 24.50, 1),
    ('Doohickey', 4.75, 12),
]


print_receipt(items)    
```

# Structural pattern matching

For *YEARS*, people new to Python complained that it didn't have a switch-case statement. And then, Python added match-case, and said that you can use this instead of `if`. 

Except that it's not the same:

- `switch-case` in C is for `if`-`else` decision-making
- `match`-`case` in Python is for breaking apart data structures without having to work hard

```python
name = 'Reuven'

match name:
    case 'Reuven':
        print('Hi, boss!')
    case 'someone else':
        print('Hi, someone else')
    case _:
        print('This is weird')
```

Traditionally, the final `case` is put into a variable. That means: It'll match anything, anything at all. The `_` variable is used for throwaway data.

```python
name = 'Reasdfsadfasdfdasdfdadsuven'

match name:
    case 'Reuven':
        print('Hi, boss!')
    case 'someone else':
        print('Hi, someone else')
    case other_name:  # the variable here is assigned the value in the "match" statement!
        print('This is weird')
```

```python
other_name
```

```python
# structural pattern matching is really meant for picking apart values from a collection -- sort of like we do with unpacking

mylist = [10, 20]

match mylist:
    case [x]:
        print(f'{x=}')
    case [x, y]:
        print(f'{x=}, {y=}')
    case [x, y, z]:
        print(f'{x=}, {y=}, {z=}')
    case other:
        print(f'{other=}')
```

```python
# we can use our unpacking *z stuff here, too:

mylist = [10, 20, 30, 40, 50, 60]

match mylist:
    case [x]:
        print(f'{x=}')
    case [x, y]:
        print(f'{x=}, {y=}')
    case [x, y, *z]:
        print(f'{x=}, {y=}, {z=}')
    case other:
        print(f'{other=}')
```

```python
# we can match dicts, too:

d = {'a':10, 'b':20, 'c':30}

match d:
    case {'a':10, 'b':20}:
        print('Partial match')
    case _:
        print('Default')
```

```python
name = 'asdfsafdsdafsafsafsafsafas'

match name:
    case x if len(x) < 5:
        print(f'Short name, {name}!')
    case x if len(x) > 10:
        print(f'Long name, {name}!')
    case _: 
        print(f'Medium name, {name}!')
```

# Modern dicts

For many years, one of the most important things to tell people about dicts was that they didn't preserve their order. If you created `{'a':10, 'b':20, 'c':30}`, you didn't know if iterating would give you `a`, `b`, `c`, or maybe `b`, `c`, and `a`.

Then, in Python 3.6 and 3.7, they changed this entirely: In 3.6, they changed the implementation of dicts, and in 3.7 they said that a nice side feature (namely, the ordering) was guaranteed.

What did they change?

A dict is a key-value store, a hash table. This means that if you store `d['a'] = 10`, Python runs the `hash('a')`, gets a number, and (in old versions of Python) would use that number to know where to store your value. Iterating over a dict meant going through each of the keys, invoking `hash` on each key, and retrieving each value.

Not only did this not guarantee the order, but it also meant that the dict needed a *lot* of space in memory. If we didn't use lots of memory, then we would end up with many colissions, which is a problem in dicts.

In 3.6, they changed the implementation to use 2 data structures:

1. a table with three columns that grows over time. This is where the key and value are stored, sequentially. The third column is the index in the table, starting with 0. There's one row in this table for each key-value pair in the dict.
2. an array (at the C level), which is where the hash function is really invoked. This is much smaller than the original dict ever was, because it's just integers in C.

If you store `d['a'] = 10`, Python first puts a new row in the table with the index (automatically calculated) and then the key-value pair. Then Python calculates `hash('a') % len(array)`, and gets an indication of where to put the index from the table.

So the array contains table-row indexes. Each lookup in the dict requires two lookups -- one in the array and one in the dict.

But we save a TON of memory! 

Not only does this save memory, but we also have the table with our key-value pairs in chronological order!

What happens, though, when we remove a key-value pair? 

Remember that all variables in Python are actually keys in a dict. Remember that every attribute (i.e., names after `.`) are really dicts. Remember that objects are really dicts.

```python
d = {'a':10, 'b':20, 'c':30}

for key, value in d.items():
    print(f'{key}: {value}')
```

```python
d.pop('b')
d['b'] = 999

for key, value in d.items():
    print(f'{key}: {value}')
```

# What else has changed?

We have new dict operators!

```python
d1 = {'a':10, 'b':20, 'c':30}
d2 = {'b':200, 'd':400}

# we could always say:

d1.update(d2)  # this means: let d2 overwrite d1
d1
```

```python
# in Python 3.9 , we have the | operator, which does the same, but non-destructively:

d1 = {'a':10, 'b':20, 'c':30}
d2 = {'b':200, 'd':400}

d1 | d2
```

# Comprehensions

Comprehensions are still a major part of Python -- and in fact, we probably use them even more!

- List comprehensions
- Dict comprehensions
- Set comprehensions
- Generator expressions

One big change in modern Python is that comprehensions *don't* leak their variables to the outside world!

```python
x = 100

[x**2 
 for x in range(5)]
```

```python
x
```

Normally, if we don't have a function, then we don't have a local scope. Does this mean that a comprehension is really a function in disguise?

The answer is: Yes, it used to be! For the first part of Python 3's life, a comprehension did create an anonymous, private function that gave it locally scoped variables.

That is no longer the case, but the variables are still locally scoped.

```python
# this is not the case with regular for loops:

x = 100

for x in range(5):
    print(x**2)

print(x)    
```
